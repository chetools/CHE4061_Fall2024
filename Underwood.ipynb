```python
import numpy as np
from scipy.optimize import root_scalar
from plotly.subplots import make_subplots
```

```python
alphas = np.array([3., 2., 1., 0.5])
zs = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
def UW(theta):
    return np.sum(alphas*zs/(alphas - theta))
```

```python
root_scalar(UW, bracket=(1.01, 1.99))
```

```python
thetaplot = np.linspace(0.4,3.1, 200)
UW_value = [UW(theta) for theta in thetaplot]
```

```python
fig = make_subplots()
fig.add_scatter(x=thetaplot, y=UW_value)
fig.update_layout(width=600,height=400,template='plotly_dark')
```

```python
Ftot = 100
zF = zs*Ftot  # mol/s
```

```python
recD = np.array([1., .99, 0.02, 0.])
```

```python
V=np.sum(recD*zF*alphas/(alphas - 1.263840561373371))  #vapor flow at minimum reflux in rectifying section
```

```python
R  = V /np.sum(zF*recD) - 1
```
